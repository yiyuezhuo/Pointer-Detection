```python
def batch_callback(frame=None, boxes=None, img_path=None, classIds=None, **kwargs):
    _,fname = os.path.split(img_path)
    debug_path = os.path.join('debug', fname)
    debug_raw_path = os.path.join('debug_raw', fname)
    if 0 not in classIds or 1 not in classIds:
        save_box_fig(frame, boxes, debug_path, figsize=(16,9))
        shutil.copy(img_path, debug_raw_path)
```

```python
batch_detect('data/*.jpg', 'results_detect_YOLO19000', savefig=True, callback = batch_callback, figsize=(16,9))
```
