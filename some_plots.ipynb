```python
cv.imwrite('merged_img.jpg', merged_img)
```

```python
width,height = img.shape[1],img.shape[0]

rows = 6
cols = 5
scale = 2

merged_img = np.empty((rows*height//scale, cols*width//scale//2, 3), dtype=np.uint8)

template = 'mini_results2/20180412145714_{}.jpg'
for idx,degree in enumerate(degrees):
    path = template.format(degree)
    img = cv.imread(path)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    
    i,j = idx//5,idx%5
    
    hs = height//scale
    ws = width//scale//2
    
    merged_img[i*hs:(i+1)*hs, j*ws:(j+1)*ws] = cv.resize(img[:,width//4:width//4*3], (ws, hs))
```

```python
cv.imwrite('merged_img_reduced.jpg', merged_img)
```
